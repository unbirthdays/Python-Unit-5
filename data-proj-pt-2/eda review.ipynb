- distribution
> the probabilities of how values are distributed within a dataset
- variance
> measures the average dispersion of a set of data points around the mean value, (summation of: (value- mean)^2)/N
- standard deviation
> measurement of dispersion of a set of data points around the mean value in original units (summation of: (value - mean)^2)/(n-1)

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
salary_numeric = pd.read_csv("salary_numeric.csv")
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = Salary and education is positively correlated (correlation coefficient is positive).

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(salary_numeric.salary, salary_numeric.age)
```

The correlation is positive.

#### Are these results statistically significant? How do you know?

P-value = 0 therefore null hypothesis is rejected. There is a correlation that is statistically significant. 

### What is the correlation between salary and age?

#### Your Hypothesis = Salary and age is positively correlated (correlation coefficient is positive).

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(salary_numeric.salary, salary_numeric.age)
```

The correlation is positive.

#### Are these results statistically significant? How do you know? 

P-value = 0 therefore null hypothesis is rejected. There is a correlation that is statistically significant. 

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = Salary and industry experience is positively correlated (correlation coefficient is positive).

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(salary_numeric.salary, salary_numeric.industry_experience)
```

The correlation is positive.

#### Are these results statistically significant? How do you know?

P-value = 0 therefore null hypothesis is rejected. There is a correlation that is statistically significant.
